```python
df["date"] = pd.to_datetime(df["date"], format = "%d-%m-%y")
```

```python
df.info()
```

```python
df["city"] = df["city"].fillna(df["city"].mode()[0])
```

```python
pd.isnull(df).sum()
```

```python
df["player_of_match"] = df["player_of_match"].fillna("No")
df["umpire1"] = df["umpire1"].fillna("No umpire")
df["umpire2"] = df["umpire2"].fillna("No umpire")
```

```python
pd.isnull(df).sum()
```

```python
df = df.dropna(subset = ["winner"])
```

```python
pd.isnull(df).sum()
```

```python
df.drop(['umpire3'], axis=1, inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
# Bar Chart

df['Season'].value_counts().sort_index().plot(kind='bar', color='steelblue')
plt.title("Bar Chart: Matches per Season")
plt.xlabel("Season")
plt.ylabel("Matches")
```

```python
 #Pie Chart

df['toss_decision'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff'])
plt.title("Pie Chart - Toss Decissions")

```

```python
# Donut Chart

winners = df['winner'].value_counts().head(5)
plt.pie(winners, labels=winners.index, autopct='%1.0f%%', startangle=140, wedgeprops=dict(width=0.2))
plt.title("Donut Chart: Top 5 Winner Teams")
```

```python
# Scatterplot

sns.scatterplot(x='win_by_runs', y='win_by_wickets', data=df, hue='winner')
plt.title("Scatter plot: Runs vs Wickets")
```

```python
# Violin Plot

import warnings
warnings.filterwarnings('ignore')
sns.violinplot(x='winner', y='win_by_wickets', data=df, palette='Set2')
plt.title("violin Plot: Wickets by winner")
```

```python
# Box Plot

sns.boxplot(x='winner', y='win_by_runs', data=df, palette = 'pastel')
plt.title("Box Plot: Win by Runs by winner")
```

```python
# KDE plot

sns.kdeplot(df['win_by_runs'], shade=True, color='magenta')
plt.title("KDE plot: Win by Runs")
```
